```python
i = 0
avg = 0
Salary_P = 101000
Salary_S = 150000
Salary_total = Salary_P+Salary_S
PS = Salary_total/Salary_P
SS = Salary_total/Salary_S
Swati_List = list()
Pratik_List = list()
def Swati_entries(): 
    while True:
        i = input("Enter amount spent by Swati: or enter done to finish the entries for Swati").lower()
        if i =="done":
            print(Swati_List)
            Swati_expense = sum(Swati_List)
            print(f"Total amount spent by Swati is",Swati_expense)
            break
        else:
            if i.isdigit():
                i = int(i)
                Swati_List.append(i)
            else:
                print("Please enter a numeric value")
                continue
#----------------------------------------------------------------------------------------------
def Pratik_entries():
    while True:
        i = input("Enter amount spent by Pratik: or enter done to finish the entries for Pratik").lower()
        if i =="done":
            print(Pratik_List)
            Pratik_expense = sum(Pratik_List)
            print(f"Total amount spent by Pratik is",Pratik_expense)
            break
        else:
            if i.isdigit():
                i = int(i)
                Pratik_List.append(i)
            else:
                print("Please enter a numeric value")
                continue
#----------------------------------------------------------------------------------------------
while True:
    print("Please Select the person you want to enter the expenses of: \n 1. Pratik\n 2. Swati\n 3. Enter 'Done' to finish the entries")
    person = input().lower()
    if person == "1" or person == "pratik":
        Pratik_entries()
    elif person == "2" or person == "swati":
        Swati_entries()
    elif person == "done":
        break
    else:
        print("Please enter a valid input")
        continue
#----------------------------------------------------------------------------------------------
print(f"List of Swati's expenses\n{Swati_List}\n Total expenditure of Swati is: {sum(Swati_List)}\n")
print(f"List of Pratik's expenses\n{Pratik_List}\n Total expenditure of Pratik is: {sum(Pratik_List)}\n")
print("\n\n")
#-----------------------------------------------------------------------------------------------
Swati_total = sum(Swati_List)
Pratik_total =sum(Pratik_List)
total = Swati_total + Pratik_total
Swati_actual = total/SS
Pratik_actual = total/PS
print(f"Total Expenditure is {total}")
print(f"Swati's share is {Swati_actual}")
print(f"Pratik's share is {Pratik_actual}")
if str(Swati_total) < str(Swati_actual):
    Swati_Diff = round((Swati_actual-Swati_total),None)
    print(f"Swati owes {Swati_Diff} to Pratik")
else:
    Pratik_Diff = round((Pratik_actual-Pratik_total),None)
    print(f"Pratik owes {Pratik_Diff} to Swati")
print("Thank you")
```
